```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
plt.style.use('seaborn')
sns.set_style('whitegrid')
%matplotlib inline
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

pd.options.mode.chained_assignment = None # Warning for chained copies disabled

def msv1(data, thresh=20, color='black', edgecolor='black', width=15, height=3):
    """function for display a graph"""
    
    plt.figure(figsize=(width,height))
    percentage=(data.isnull().mean())*100
    percentage.sort_values(ascending=False).plot.bar(color=color, edgecolor=edgecolor)
    plt.axhline(y=thresh, color='r', linestyle='-')
    plt.title('Valeurs manquantes par colonnes', fontsize=20, weight='bold' )
    plt.text(len(data.isnull().sum()/len(data))/1.7, thresh+12.5, 'colonnes avec plus de %s%s de valeurs manquantes' %(thresh, '%'), fontsize=12,weight='bold', color='crimson',ha='left' ,va='top')
    plt.text(len(data.isnull().sum()/len(data))/1.7, thresh - 5, 'colonnes avec moins de %s%s de valeurs manquantes' %(thresh, '%'), fontsize=12,weight='bold', color='blue',ha='left' ,va='top')
    plt.xlabel('Colonne', size=15, weight='bold')
    plt.ylabel('pourcentage valeur manquante', weight='bold')
    plt.yticks(weight ='bold')
    
    return plt.show()
```

```python
df = pd.read_csv("/home/apprenant/simplon_project/house_price/data/train.csv")
```
